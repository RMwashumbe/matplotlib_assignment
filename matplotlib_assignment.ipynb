```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Create a figure object
fig = plt.figure()

# Add the main axis to the figure
ax1 = fig.add_axes([0, 0, 1, 1])

# Add the inset axis to the figure
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

# Plot data on the main axis
#ax1.plot(x, y, label='y = x*2')
ax1.plot(x, z, label='z = x^2')
ax1.set_xlabel('X')
ax1.set_ylabel('Y and Z')
ax1.set_title('Main Plot')
ax1.legend()

# Plot data on the inset axis
#ax2.plot(x, y, 'r', label='y = x*2')
ax2.plot(x, z, 'b', label='z = x^2')
ax2.set_xlabel('X')
ax2.set_ylabel('Y and Z')
ax2.set_title('Inset Plot')
ax2.set_xlim(20, 22)
ax2.set_ylim(0, 1000)
ax2.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot (x, y) on the first axis
ax1.plot(x, y, label='y = x*2')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Plot of y = x*2')
ax1.legend()

# Plot (x, z) on the second axis
ax2.plot(x, z, label='z = x^2', color='r')
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_title('Plot of z = x^2')
ax2.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.arange(0, 100)
y = x * 2
z = x ** 2

# Create a figure with two subplots and specify the figure size
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8, 3))

# Plot (x, y) on the first axis
ax1.plot(x, y, label='y = x*2')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Plot of y = x*2')
ax1.legend()

# Plot (x, z) on the second axis
ax2.plot(x, z, label='z = x^2', color='r')
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_title('Plot of z = x^2')
ax2.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```
